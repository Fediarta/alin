# Konsep Eigen value and Eigen vektor

### Penjelasan Konsep Eigen value

* matriks A dengan ukuran $n × n $ maka vektor tidak-nol x di $R^n $ disebut *Vektor eigen* dari A jika Ax sama dengan perkalian suatu skalar $\lambda$ dengan *x* , yaitu

$ Ax = \lambda x$

 Ax : vektor x setelah transformasi linier

 $ λx $: perkalian bilangan riel dari x , i.e. peregangan dari vektor x. Transformasi linier untuk eigenvektor menghasilkan peregangan vektor sebanyak eigen value kali.


skalar $λ$ disebut *nilai eigen* dari A , dan *x* dinamakan vektor eigen yang berkoresponden dengan  $λ$ / nilai eigen


*nb: vektor nol tidak bisa menjadi vektor eigen meskipun A0 = $λ0$,namun $λ$= 0 bisa menjadi nilai eigen*

### jika ($λ$I-A)=0 memiliki solusi tidak-nol , maka haruslah

*det ($λ$I-A)=0*

* persamaan det ($λ$I-A)=0 disebut persamaan karakteristik dari matriks A dan akar-akar persamaan tersebut , yaitu $λ$,dinamakan akar-akar karakteristik atau nilai-nilai eigen .

# contoh menyelesaikan eigen value dan aigen vektor dengan library numpy


```python
import numpy as np
from numpy.linalg import eig

a = np.array([[1, 4],
              [2, 3]])

print(a)
w,v=eig(a)
print('E-value:', w)
print('E-vector', v)
```

# implementasi dari code yang dikirim (hanya mengeluaran matriks dari proses tsb)

```python
import numpy as np


def pcaEVD(X):
    # Center X and get covariance matrix C
    n, p = X.shape

    X -= X.mean(axis=0)
    C  = (X.T@X) / (n-1)




    # EVD
    e_vals, e_vecs = np.linalg.eig(C)
    idx = e_vals.argsort()[::-1]
    e_vals = e_vals[idx]
    e_vecs = e_vecs[:, idx]


    # Return principal compnents and eigenvalues to calculate the portion of sample variance explained
    return (X @ e_vecs), e_vals




```

```python
def pcaSVD(X):
    # Center X and get covariance matrix C
    n, p = X.shape
    X -= X.mean(axis=0)

    # SVD
    u, sigma, vt = np.linalg.svd(X, full_matrices=False)




    # Return principal compnents and eigenvalues to calculate the portion of sample variance explained
    return (X @ vt.T), (sigma**2) / (n-1)

print("U matrix:")
print(u)
print("\nSigma matrix:")
print(np.diag(sigma))
print("\nV^T matrix:")
print(vt)
```

```python
import numpy as np
from sklearn.datasets import load_iris

# Memuat dataset Iris
iris = load_iris()
X = iris['data']
y = iris['target']
print("X=\n",X)
print("Y=\n",y)

# Sentralisasi data
X_ = X - X.mean(axis=0)
print("Sentralisasi data=\n",X_)

# SVD
u, sigma, vt = np.linalg.svd(X_, full_matrices=False)

# Komponen utama
principal_components = X_ @ vt.T

# Nilai eigen
eigenvalues = (sigma**2) / (X.shape[0] - 1)

# Mencetak hasil
print("Principal Components:\n", principal_components)
print("Eigenvalues:\n", eigenvalues)



```
