 # PageRank
 -------------------------

 Algoritma pagerank pertama kali diperkenalkan oleh lary page.biasanya digunakan untuk menetukan peringkat halaman web dimesin pencarian google,dan menentukan peringkat pengguna di media sosial . sehingga dengan algoritma pagerank kita mampu memperkirakan seberapa bagus suatu situs web.


 ## Cara Kerja PageRank untuk menghitung nilai web:
 -------------------------

1. **Jumlah Tautan**: Semakin banyak tautan yang mengarah ke suatu halaman, semakin penting halaman tersebut.
2. **Kualitas Tautan**: Tautan dari halaman dengan PageRank tinggi memiliki bobot yang lebih tinggi dibandingkan tautan dari halaman dengan PageRank rendah.
3. **Konektivitas Link**: Distribusi link dan konektivitasnya di setiap situs web juga mempengaruhi PageRank.


PageRank bekerja dengan menghitung jumlah dan kualitas tautan ke suatu halaman untuk menentukan perkiraan kasar seberapa penting situs web tersebut. Algoritma ini memberikan nilai mulai dari 0 hingga 10 untuk menunjukkan seberapa berkualitasnya website tersebut. Semakin tinggi nilai PageRank, semakin kredibel website tersebut dianggap.

### Faktor Penentu PageRank
-------------------------

Beberapa faktor yang mempengaruhi PageRank adalah:

1. **Backlinks**: Jumlah dan kualitas tautan masuk yang mengarah ke suatu halaman.
2. **Internal Linking**: Tautan internal yang mengarah ke halaman lain di dalam situs web.
3. **External Linking**: Tautan eksternal yang mengarah ke situs web lain.



### Status PageRank Saat Ini
-------------------------

Walaupun PageRank tidak lagi ditampilkan secara publik, algoritma ini masih digunakan dalam sistem internal Google. Namun, penting untuk diingat bahwa PageRank bukanlah satu-satunya faktor yang menentukan ranking halaman di hasil pencarian Google. Ada banyak faktor lain yang juga berperan, seperti konten halaman, kegunaan, kecepatan loading, dan lain-lain.



Rumus PageRank yang asli dikembangkan oleh pendiri Google Lawrence Page dan Sergey Brin adalah sebagai berikut:
$
PR(A) = (1 - d) + d (PR(T1) / C(T1) + ... + PR(Tn) / C(Tn))
$
Di mana:

- PR(A) adalah PageRank halaman A
- PR(Ti) adalah PageRank halaman Ti yang menautkan ke halaman A
- C(Ti) adalah jumlah tautan eksternal halaman Ti
- d adalah koefisien dump yang berkisar dari 0 hingga 1, biasanya nilai acak adalah 0,85[3].

```python
import numpy as np

A = np.array([[0, 0, 1, 1/2],
              [1/3, 0, 0, 0],
              [1/3, 1/2, 0, 1/2],
              [1/3, 1/2, 0, 0]])

v = np.array([[0.25], [0.25], [0.25], [0.25]])
#v = np.array([[0.20], [0.30], [0.20], [0.30]])

previous_result = v
tolerance = 0.01  # Ubah nilai toleransi menjadi 0.01

for i in range(1, 100):  # Batasi jumlah iterasi untuk mencegah perulangan tak terbatas
    result = A @ previous_result
    print(f"(A^{i})v = \n", result)

    # Cek kondisi berhenti
    if np.all(np.abs(result - previous_result) <= tolerance):
        if np.sqrt(np.sum((result - previous_result) ** 2)) < tolerance:
          print(f"\nPerkalian matriks konvergen pada iterasi ke-{i}.")
          break

    previous_result = result
```
