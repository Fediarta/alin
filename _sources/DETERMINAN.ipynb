# DETERMINAN

### contoh fungsi untuk menghitung determinan dalam python



```python
import numpy as np
A = np.array([[6,2],[4,2]])
B = np.array([[3,7],[4,8]])
C = np.array([[1,0],[0,1]])
D = np.array([[1,0,0],[0,1,0],[0,0,1]])
F = np.array([[1,4,2],[1,1,1],[3,2,1]])



det_A = np.linalg.det(A)
det_B = np.linalg.det(B)
det_C = np.linalg.det(C)
det_D = np.linalg.det(D)
det_F = np.linalg.det(F)




print("Matriks 2x2")
print(f"matrik A:\n {A}")
print(f"Determinan Matrix A :\n{det_A}")
print("")
print(f"matrik B:\n{B}")
print(f"Determinan Matrix B :\n{det_B}")
print("")
print(f"matrik C:\n{C}")
print(f"Determinan Matrix C :\n{det_C}")
print("")
print("Matriks 3x3  ")
print(f"matrik D:\n{D}")
print(f"Determinan Matrix D :\n{det_D}")
print("")
print(f"matrik F:\n{F}")
print(f"Determinan Matrix F :\n{det_F}")
print("")


inv_A = np.linalg.inv(A)
inv_B = np.linalg.inv(B)
print(f"invers Matrix A :\n{inv_A}")
print(f"invers Matrix B :\n{inv_B}")
```

# Sifat-Sifat Derterminan Matriks
4.jika matriks adalah kelipatan sebanyak n * matrix maka untuk mencari determinan adalah (n^(sebanyak ordo)determinan matrix)contoh matrix i adalah kelipatan 2 kali dari matrix G : (2^3(det G))



```python
import numpy as np

i = np.array([[4, 4, 6], [2, 6, 2], [4, 8, 4]])
det_i = np.linalg.det(i)
rounded_det_i = round(det_i)

print(f"Matriks i:\n{i}")
print(f"Determinan Matriks i:\n{rounded_det_i}")
```
