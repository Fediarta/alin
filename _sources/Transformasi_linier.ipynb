# TRANSFORMASI LINIER
definisi 1
Transformasi Linier adalah misalkan V dan W adalah ruang vektor . Transformasi yang memetakan ruang vektor V ke ruang vektor w ditulis sebagai berikut

 $T:V==>W$

 V adalah daerah asal (domain) transformasi T dan W adalah daerah hasil
transformasi (kodomain) fungsi. Jika V = W, maka T dinamakan operator
pada V.

$Jika, v ∈ V dan w ∈ W maka,
w = T(v)$





definisi 2

Misalkan V dan W adalah ruang vektor.Transformasi

$T:V==>W$

dinamakan transformasi linier jika untuk semua u dan v di dalam V dan
k sebuah skalar berlaku:

$1.T(u+v)=T(u)+T(v)$

$2.T(ku)=kT(u)$

Jika V = W, maka T dinamakan operator linier pada V.

## contoh yang membuktikan rumus 1
misalkan
 $\overline{\rm U}=
\begin{pmatrix}
U1 \\
U2
\end{pmatrix}$,
$\overline{\rm V}=
\begin{pmatrix}
V1 \\
V2
\end{pmatrix}$

akan ditunjukkan bahwa
$T(u+v)=T(u)+T(v)$

$T(u+v)=T$
$\left(
\begin{pmatrix}
U1 \\
U2
\end{pmatrix}+,
\begin{pmatrix}
V1 \\
V2
\end{pmatrix}
 \right)=$
$\begin{pmatrix}
(u1+v1)-(u2+v2) \\
-(u1+v1) \\
(u2+v2)
\end{pmatrix}=$
$\begin{pmatrix}
u1+v1-u2-v2 \\
-u1-v1 \\
u2+v
\end{pmatrix}=$
$\begin{pmatrix}
u1-v1+u2+v2 \\
-u1-v1 \\
u2+v2
\end{pmatrix}=$
$\begin{pmatrix}
u1-u2 \\
-u1\\
u2
\end{pmatrix}+$
$\begin{pmatrix}
v1-v2 \\
-v1\\
v2
\end{pmatrix}=$
$T(\overline{\rm U})+T(\overline{\rm V})$


Terbukti bahwa $T(\overline{\rm U})+T(\overline{\rm V})=$
$T(\overline{\rm U})+T(\overline{\rm V})$


```python
import numpy as np

# Definisikan transformasi linier sebagai perkalian dengan matriks
def linear_transformation(T, v):
    return np.dot(T, v)

# Matriks transformasi contoh
T = np.array([[2, 0],
              [0, 3]])

# Vektor contoh
u = np.array([1, 2])
v = np.array([3, 4])

# Hitung T(u + v)
lhs = linear_transformation(T, u + v)

# Hitung T(u) + T(v)
rhs = linear_transformation(T, u) + linear_transformation(T, v)

# Periksa apakah keduanya sama
print("T(u + v):", lhs)
print("T(u) + T(v):", rhs)
print("Apakah T(u + v) == T(u) + T(v)?", np.array_equal(lhs, rhs))
```

## Contoh untuk membuktikan bahwa
## T(ku)=kT(u)

$T(ku)=T$
$\left(
k\begin{pmatrix}

$\begin{pmatrix}

$a\begin{pmatrix}

u2
\end{pmatrix}=$
$kT(\overline{\rm U})$

Terbukti bahwa
$T(k \overline{\rm U})=$
$kT( \overline{\rm U})$


```python
import numpy as np

# Definisikan transformasi linier sebagai perkalian dengan matriks
def linear_transformation(T, v):
    return np.dot(T, v)

# Matriks transformasi contoh
T = np.array([[2, 0],
              [0, 3]])

# Vektor contoh
u = np.array([1, 2])

# Skalar contoh
c = 5

# Hitung T(c * u)
lhs = linear_transformation(T, c * u)

# Hitung c * T(u)
rhs = c * linear_transformation(T, u)

# Periksa apakah keduanya sama
print("T(c * u):", lhs)
print("c * T(u):", rhs)
print("Apakah T(c * u) == c * T(u)?", np.array_equal(lhs, rhs))
```

# Refleksi pada Transformasi

Refleksi adalah jenis transformasi linier yang memetakan setiap titik pada suatu ruang ke titik yang simetris terhadap suatu garis, bidang, atau hiperpilan tertentu. Dalam konteks dua dimensi, refleksi biasanya dilakukan terhadap garis (misalnya, sumbu x atau sumbu y), sedangkan dalam tiga dimensi, refleksi dilakukan terhadap bidang (misalnya, bidang xy, yz, atau zx).

### Sifat -sifat Refleksi pada Transfromasi
* Bangun (objek) yang dicerminkan (refleksi) tidak mengalami perubahan bentuk dan ukuran .

* Jarak bangun (objek) dari cermin (cermin datar) adalah sama dengan jarak bayangan dengan cermin tersebut


### Refleksi di Dua Dimensi

1. **Refleksi terhadap sumbu x:**
   - Matriks refleksi terhadap sumbu x adalah:
     \[
     R_x = \begin{pmatrix}
     1 & 0 \\
     0 & -1
     \end{pmatrix}
     \]
   - Transformasi ini mengubah titik \( (x, y) \) menjadi \( (x, -y) \).

2. **Refleksi terhadap sumbu y:**
   - Matriks refleksi terhadap sumbu y adalah:
     \[
     R_y = \begin{pmatrix}
     -1 & 0 \\
     0 & 1
     \end{pmatrix}
     \]
   - Transformasi ini mengubah titik \( (x, y) \) menjadi \( (-x, y) \).

3. **Refleksi terhadap garis \( y = x \):**
   - Matriks refleksi terhadap garis \( y = x \) adalah:
     \[
     R_{y=x} = \begin{pmatrix}
     0 & 1 \\
     1 & 0
     \end{pmatrix}
     \]
   - Transformasi ini mengubah titik \( (x, y) \) menjadi \( (y, x) \).



```python
import numpy as np

# Definisikan fungsi refleksi
def reflection(matrix, point):
    return np.dot(matrix, point)

# Matriks refleksi terhadap sumbu x
R_x = np.array([[1, 0],
                [0, -1]])

# Matriks refleksi terhadap sumbu y
R_y = np.array([[-1, 0],
                [0, 1]])

# Matriks refleksi terhadap garis y=x
R_y_equals_x = np.array([[0, 1],
                         [1, 0]])

# Titik contoh
point = np.array([1, 2])

# Refleksi terhadap sumbu x
reflected_x = reflection(R_x, point)
print("Refleksi terhadap sumbu x:", reflected_x)

# Refleksi terhadap sumbu y
reflected_y = reflection(R_y, point)
print("Refleksi terhadap sumbu y:", reflected_y)

# Refleksi terhadap garis y=x
reflected_y_equals_x = reflection(R_y_equals_x, point)
print("Refleksi terhadap garis y=x:", reflected_y_equals_x)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definisikan fungsi refleksi
def reflection(matrix, point):
    return np.dot(matrix, point)

# Matriks refleksi terhadap sumbu x
R_x = np.array([[1, 0],
                [0, -1]])

# Matriks refleksi terhadap sumbu y
R_y = np.array([[-1, 0],
                [0, 1]])

# Matriks refleksi terhadap garis y=x
R_y_equals_x = np.array([[0, 1],
                         [1, 0]])

# Titik contoh
point = np.array([1, 2])

# Refleksi terhadap sumbu x
reflected_x = reflection(R_x, point)
# Refleksi terhadap sumbu y
reflected_y = reflection(R_y, point)
# Refleksi terhadap garis y=x
reflected_y_equals_x = reflection(R_y_equals_x, point)

# Plot titik asli dan hasil refleksi
plt.figure(figsize=(8, 8))
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.gca().set_aspect('equal', adjustable='box')

# Garis y=x
x_vals = np.array(plt.gca().get_xlim())
y_vals = x_vals
plt.plot(x_vals, y_vals, '--', color='blue', label='Garis y=x')

# Plot titik
plt.plot(point[0], point[1], 'o', label='Titik Asli (1, 2)', color='red')
plt.plot(reflected_x[0], reflected_x[1], 'o', label='Refleksi terhadap sumbu x', color='green')
plt.plot(reflected_y[0], reflected_y[1], 'o', label='Refleksi terhadap sumbu y', color='purple')
plt.plot(reflected_y_equals_x[0], reflected_y_equals_x[1], 'o', label='Refleksi terhadap garis y=x', color='orange')

# Anotasi
plt.annotate('Asli', (point[0], point[1]), textcoords="offset points", xytext=(5,-10), ha='center')
plt.annotate('Refleksi x', (reflected_x[0], reflected_x[1]), textcoords="offset points", xytext=(5,-10), ha='center')
plt.annotate('Refleksi y', (reflected_y[0], reflected_y[1]), textcoords="offset points", xytext=(5,-10), ha='center')
plt.annotate('Refleksi y=x', (reflected_y_equals_x[0], reflected_y_equals_x[1]), textcoords="offset points", xytext=(5,-10), ha='center')

# Menampilkan legenda
plt.legend()

# Menampilkan plot
plt.show()
```

### Refleksi di Tiga Dimensi

1. **Refleksi terhadap bidang xy:**
   - Matriks refleksi terhadap bidang xy adalah:
     \[
     R_{xy} = \begin{pmatrix}
     1 & 0 & 0 \\
     0 & 1 & 0 \\
     0 & 0 & -1
     \end{pmatrix}
     \]
   - Transformasi ini mengubah titik \( (x, y, z) \) menjadi \( (x, y, -z) \).

2. **Refleksi terhadap bidang yz:**
   - Matriks refleksi terhadap bidang yz adalah:
     \[
     R_{yz} = \begin{pmatrix}
     -1 & 0 & 0 \\
     0 & 1 & 0 \\
     0 & 0 & 1
     \end{pmatrix}
     \]
   - Transformasi ini mengubah titik \( (x, y, z) \) menjadi \( (-x, y, z) \).

3. **Refleksi terhadap bidang zx:**
   - Matriks refleksi terhadap bidang zx adalah:
     \[
     R_{zx} = \begin{pmatrix}
     1 & 0 & 0 \\
     0 & -1 & 0 \\
     0 & 0 & 1
     \end{pmatrix}
     \]
   - Transformasi ini mengubah titik \( (x, y, z) \) menjadi \( (x, -y, z) \).


!

```python
# contoh code untuk refleksi 3 dimensi
import numpy as np

# Definisikan fungsi refleksi
def reflection(matrix, point):
    return np.dot(matrix, point)

# Matriks refleksi terhadap bidang xy
R_xy = np.array([[1, 0, 0],
                 [0, 1, 0],
                 [0, 0, -1]])

# Matriks refleksi terhadap bidang yz
R_yz = np.array([[-1, 0, 0],
                 [0, 1, 0],
                 [0, 0, 1]])

# Matriks refleksi terhadap bidang zx
R_zx = np.array([[1, 0, 0],
                 [0, -1, 0],
                 [0, 0, 1]])

# Titik contoh
point = np.array([3, 4, 5])

# Refleksi terhadap bidang xy
reflected_xy = reflection(R_xy, point)
print("Refleksi terhadap bidang xy:", reflected_xy)

# Refleksi terhadap bidang yz
reflected_yz = reflection(R_yz, point)
print("Refleksi terhadap bidang yz:", reflected_yz)

# Refleksi terhadap bidang zx
reflected_zx = reflection(R_zx, point)
print("Refleksi terhadap bidang zx:", reflected_zx)
```
